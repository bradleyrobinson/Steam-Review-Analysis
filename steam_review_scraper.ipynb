```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
games = pd.read_csv("games.csv")
small_dataset = games.iloc[0:10]
small_dataset
```

```python
rise_id = small_dataset['game_id'].iloc[0]
```

```python
response = requests.get("http://steamcommunity.com/app/" + str(rise_id) + "/reviews/?p=1&browsefilter=mostrecent")
content  = response.content
parser = BeautifulSoup(content, 'html.parser')
```

```python
parser
```

```python
divs = parser.find_all('div')
reviews = {'recommended': [], 'text': []}
for d in divs:
    attribs = d.attrs
    if 'class' in attribs:
        if attribs['class'] == ['title']:
            reviews['recommended'].append(d.text)
        elif attribs['class'] == ['apphub_CardTextContent']:
            reviews['text'].append(d.text)
print(len(reviews['recommended']), len(reviews['text']))
```

So now that it's working, it's time to see if we can look at more pages for this particular game...

```python
def get_reviews(divs):
    reviews = {'recommended': [], 'text': []}
    for d in divs:
        attribs = d.attrs
        if 'class' in attribs:
            if attribs['class'] == ['title']:
                reviews['recommended'].append(d.text)
            elif attribs['class'] == ['apphub_CardTextContent']:
                reviews['text'].append(d.text)
    return reviews

def get_page(page, steam_id):
    response = requests.get("http://steamcommunity.com/app/" + str(steam_id) + "/reviews/?p=" + str(page) +"&browsefilter=mostrecent")
    content  = response.content
    parser = BeautifulSoup(content, 'html.parser')
    divs = parser.find_all('div')
    reviews = get_reviews(divs)
    return reviews
```

```python
reviews = []
for i in range(0, 100):
    try:
        reviews.append(get_page(i, rise_id))
    except:
        print('Done!')
```
